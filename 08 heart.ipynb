```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
# Line Plot
plt.figure(figsize=(15,10))
sns.lineplot(df['trtbps'])
```

```python
# Bar Plot

plt.figure(figsize=(15,10))
sns.barplot(x=df['age'], y=df['trtbps'])
```

```python
# Histogram

sns.histplot(df['age'])
```

```python
#violin plot

sns.violinplot(df['age'])
```

```python
# Box Plot

plt.figure(figsize=(15,10))
sns.boxplot(x=df['age'], y=df['trtbps'])
```

```python
#histogram plot

sns.distplot(df['age'])
```

```python
#Scatter plot

sns.scatterplot(x=df['age'], y=df['trtbps'])
```

```python
#heat Map
sns.heatmap(df.corr())
```

```python
#Tree Map

import squarify
squarify.plot(df['trtbps'])
```
